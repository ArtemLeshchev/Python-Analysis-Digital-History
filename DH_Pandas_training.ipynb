```python
import pandas as pd
```

```python
# Создадаим структуру данных Датафрейм (далее - дф)
# Дф - словарь, где ключ = названия колонки (она же структура данных Серия), а значения = список значений этой колонки
df = pd.read_excel('DH_source_data.xlsx')
```

```python
# Пказать данные из 1 колонки (т.е. выбрать pandas Series). Для этого нужно фильтровать по колонкам 
# Поскольку дф это словарь, то необходимо, выбрать ключ словаря, чтобы отобразить значения.
# Как фильтровать по рядам - будет описано дальше.
df['Likes'] 
```

```python
# Как выбрать дф из >=2 колонок. 
# Используем 4 квадратные скобки. [] это selection brackets, а другие [] это список ключей
likes_reposts_comments = df[["Likes", "Comments", "Reposts"]]
likes_reposts_comments
```

```python
# Посмотрим средние и медианные значения
print(likes_reposts_comments.median())
print(likes_reposts_comments.mean())
```

```python
# Напишем условие с оператором сравнения. 
# Для этого отфильтруем дф по колонке и проверим, удовлетворяют ли условию значения каждого ряда или нет.
# Выведем результат да/нет для каждой строчки на соответствие условию:
df["Reposts"] > 100
```

```python
# Как фильтровать ряды.
# Чтобы отфильтровать ряды датафрейма (в моем случае - исходного), нужно в [] передать условие для фильтрации.
# Посмотрим все посты, набравшие более 100 комментариев.

df[df["Reposts"] > 100]
```

```python
# Посмотреть массив уникальных значений (аналог distinct в SQL)
df["Author"].unique()
```

```python
# Посмотреть все ряды, отфильтровав по колоне Author со значением X ИЛИ значением Y. 
# isin это аналог IN() в SQL.
df[df["Author"].isin(["id163491415", "id84987784"])]
```

```python
# Проверка True/False содержат ли ряды NA-значение.
df.notna() # Датасет не содержит Null-значений.
```

```python
# Проверка True/False содержат ли ряды NA-значение.
df.isna() # Датасет не содержит Null-значений.
```

```python
# Более сложная фильтрация. Вывести все репосты, у которых боле 500 комментов и более 1000 лайков.
df.loc[(df["Comments"] > 500) & (df["Likes"] > 1000), "Reposts"]
```

```python
# Фильтрация по индексам. В первом аргументе указываются ряды, во втором колонки.

df.iloc[9:20, [1, 2, 3, 4, 6]]
```

```python
# groupby сам по себе - это объект, который содержит метаданные о том как выполниь groupby. 
# Без агрегации просто отобразится упоминание об этом
df[["Likes", "Reposts", "Comments", "Author"]].groupby("Author")
```

```python
# Применим к groupby агрегацию, чтобы выполнить манипуляцию над данными
df[["Likes", "Reposts", "Comments", "Author"]].groupby("Author").mean()
```
